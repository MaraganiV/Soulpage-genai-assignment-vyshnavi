```python
def data_collector(state):
    movie = state["movie"]
    data = {
        "name": movie,
        "rating": "7.5/10",
        "collection": "₹600 Cr",
        "reviews": "Action blockbuster with mixed story feedback"
    }
    state["data"] = data
    return state


def analyst(state):
    data = state["data"]
    summary = f"""
    Movie: {data['name']}
    Rating: {data['rating']}
    Box Office: {data['collection']}
    Insight: Strong commercial success but story got mixed reviews.
    """
    state["analysis"] = summary
    return state
```

```python
from langgraph.graph import StateGraph

builder = StateGraph(dict)

builder.add_node("collector", data_collector)
builder.add_node("analyst", analyst)

builder.set_entry_point("collector")
builder.add_edge("collector", "analyst")

graph = builder.compile()
```

```python
result = graph.invoke({"movie": "Leo"})
print(result["analysis"])
```
